```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
```

```python
df = pd.read_csv("IRIS.csv")
df
```

##Surface Plot

```python
X = df.sepal_length
Y = df.sepal_width
def getZ(X, Y):
    return np.sin(np.sqrt(X ** 2 + Y ** 2))

X, Y = np.meshgrid(X, Y)
Z = getZ(X, Y)

fig = plt.figure()
ax = Axes3D(fig)

surf = ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

ax.set_title('Surface Plot')

plt.show()
```

Puncak Plot sesuai dengan nilai Z, terjadi pada Sepal Length Yang Mendekati Nilai Tertinggi dan Sepal Widht pada range 2 - 4

##Contour Plot

```python
X = df.sepal_length
Y = df.sepal_width
def func(X, Y):
    return np.sin(X) ** 5 + np.cos(10 + Y * X) * np.cos(X)

X, Y = np.meshgrid(X, Y)
Z = func(X, Y)

fig = plt.figure()
ax = plt.axes()

ax.contour  (X, Y, Z
            , cmap = 'winter'
            );
ax.set_xlabel('x')
ax.set_ylabel('y')
plt.show()
```

Puncak Plot nilai Z yang diinterpretasi dengan warna biru->hijau, terjadi pada Sepal Length Yang Mendekati Nilai Tertinggi dan Sepal Widht pada range 2 - 4
